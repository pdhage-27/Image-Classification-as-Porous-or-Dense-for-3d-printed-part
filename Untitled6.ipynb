```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
from tqdm import tqdm
import numpy as np
import os 
from random import shuffle
import cv2
```

```python
num_train_images = 218
num_test_images = 30
```

```python
TEST_DIR = '/Users/priyadhage/Desktop/Porous-vs-NonPorous/test1'
TRAIN_DIR = '/Users/priyadhage/Desktop/Porous-vs-NonPorous/train'
LEARNING_RATE = 1e-3
MODEL_NAME = "PorousvsNonporous-{}-{}.model".format(LEARNING_RATE,"6conv-fire")
IMAGE_SIZE = 100
```

```python
def label_image(img):

    img_name = img.split(".")[0]
     
    if img_name == "Porous":
        return [1,0]
    elif img_name == "Non-Porous":
        return [0,1]
#This Function will Return a Vector for the labels of the Data
```

```python
#This is for Training Data
def train_data_loder():
    training_data = []
    for img in tqdm(os.listdir(path=TRAIN_DIR)):
        img_lable = label_image(img)
        path_to_img = os.path.join(TRAIN_DIR,img)
        img = cv2.resize(cv2.imread(path_to_img,cv2.IMREAD_GRAYSCALE),(IMAGE_SIZE,IMAGE_SIZE))
        training_data.append([np.array(img),np.array(img_lable)])
        
    shuffle(training_data)
    np.save("training_data_new.npy",training_data)
    return training_data
```

```python
#This is for Testing Data
def testing_data():
    test_data = []
    for img in tqdm(os.listdir(TEST_DIR)):
        img_labels = img.split(".")[0]
        path_to_img = os.path.join(TEST_DIR,img)
        img = cv2.resize(cv2.imread(path_to_img,cv2.IMREAD_GRAYSCALE),(IMAGE_SIZE,IMAGE_SIZE))
        test_data.append([np.array(img),np.array(img_labels)])
    shuffle(test_data)
    np.save("test_dataone.npy",test_data)
    return test_data
```

```python
import tflearn
from tflearn.layers.conv import conv_2d,max_pool_2d
from tflearn.layers.core import input_data,dropout,fully_connected
from tflearn.layers.estimator import regression
```

```python
import tensorflow as tf
tf.reset_default_graph()

convnet = input_data(shape=[None, IMAGE_SIZE, IMAGE_SIZE, 1], name='input')
#Conv Layer 1
convnet = conv_2d(convnet, 16, 5, padding= 'SAME', activation='relu')
convnet = max_pool_2d(convnet, 5, padding= 'SAME')
#Conv Layer 2
convnet = conv_2d(convnet, 32, 5, padding= 'SAME', activation='relu')
convnet = max_pool_2d(convnet, 5, padding= 'SAME')
#Conv Layer 3
convnet = conv_2d(convnet, 16, 5, padding= 'SAME', activation='relu')
convnet = max_pool_2d(convnet, 5, padding= 'SAME')
#Conv Layer 6
convnet = fully_connected(convnet, 512, activation='relu')
convnet = dropout(convnet, 0.25)
#Fully Connected Layer with SoftMax as Activation Function
convnet = fully_connected(convnet, 2, activation='softmax')
#Regression for ConvNet with ADAM optimizer
convnet = regression(convnet, optimizer='adam', learning_rate=LEARNING_RATE, loss='categorical_crossentropy', name='targets')

model = tflearn.DNN(convnet, tensorboard_dir='log')
```

```python
training_data_g = train_data_loder()
```

```python
train = training_data_g[:]
test = training_data_g[:]
#This is our Training data
X = np.array([i[0] for i in train]).reshape(-1,IMAGE_SIZE,IMAGE_SIZE,1)
Y = [i[1] for i in train]

#This is our Training data
test_x = np.array([i[0] for i in test]).reshape(-1,IMAGE_SIZE,IMAGE_SIZE,1)
test_y = [i[1] for i in test]

```

```python
model.fit(X, Y, n_epoch=10, validation_set=(test_x,  test_y), 
    snapshot_step=10, show_metric=True, run_id=MODEL_NAME)
```

```python
model.save(MODEL_NAME) 
```

```python
import matplotlib.pyplot as plt
test_data = testing_data()
fig = plt.figure()
#test_data = np.load('test_data.npy')


for num, data in enumerate(test_data[:5]):
    #  is porous 1:0, non-porous is 0:1
    img_num = data[1]
    img_data = data[0]
    img_feed = img_data.reshape(IMAGE_SIZE,IMAGE_SIZE,1)
    
    y = fig.add_subplot(4,6, num + 1)
    orig = img_data
#     print(data)

    model_out = model.predict([img_feed])[0]
    
    if np.argmax(model_out) == 1: str_label='Porous'
    else: str_label = 'Non-Porous'
        
    y.imshow(orig, cmap='gray')
    plt.title(str_label)
    y.axes.get_xaxis().set_visible(False)
    y.axes.get_yaxis().set_visible(False)
    
plt.show()
```
